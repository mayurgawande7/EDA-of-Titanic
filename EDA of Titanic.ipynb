#### Zscore()

```python
from scipy.stats import zscore 
```

```python
z=np.abs(zscore(a["fare"]))
z
```

```python
a["zscore"]=z
```

```python
a.head()
```

```python
(a["zscore"]>=3).sum()
```

```python
az=a[a["zscore"]<3]
az
```

```python
q2=a["fare"].median()
q2
```

```python
q1=a["fare"].quantile(0.25)
q1
```

```python
q3=a["fare"].quantile(0.75)
q3
```

### IQR()

```python
iqr=q3-q1
iqr
```

```python
ai=a[~((a["fare"]<(q1-1.5*iqr))|(a["fare"]>(q3+1.5*iqr)))]
ai
```

```python
sns.boxplot(ai["fare"])
```

```python
sns.distplot(ai['fare'])
plt.show()
```
